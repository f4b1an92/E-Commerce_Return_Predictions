# Feature Selection

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler
import datetime
import re

import feature_gen as fg
import plot_fcts as pf

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)

notebook_path = os.path.abspath("Feature Selection.ipynb")
repo_path = "\\".join(notebook_path.split("\\")[:-2])
data_path = repo_path + "\\data"
```

## 1. Create Feature Space

```python
arg_dict = {
    'agg_fun': ['mean', 'median', 'max']
    , 'window': 7
    , 'feature_list': ['item_id', 'user_id', 'item_color', 'item_size', 'brand_id', 'user_state']
    , 'thr': [1, 5, 1, 1, 1, 1]
    , 'correction': 0.00001 
}
fgen = fg.FeatureGen(known_filename='BADS_WS1819_known', unknown_filename='BADS_WS1819_unknown', load_path=data_path, save_path=data_path)
fgen.prep_wrapper(**arg_dict, cat_trans='WOE')

# Load prepared data sets
known = pd.read_parquet(data_path + '/known_cleaned_woe.parquet')
known.fillna(-1, inplace=True)
unknown = pd.read_parquet(data_path + '/unknown_cleaned_woe.parquet')
unknown.fillna(-1, inplace=True)
```

```python
corr_mat = pd.concat([known['return'], known.loc[:,known.columns != 'return']], axis=1).corr()

fig, ax = plt.subplots(1,1,figsize=(18,12))

sns.heatmap(data=corr_mat,ax=ax, annot=True, fmt=".2f", cbar=False)
plt.show()
```

__Variables to kick out because of colinearity:__
- __most_exp_item__ (because correlation between it and "solo_purchase" is above 0.5 and solo_purchase has a higher correlation with target)
- __rel_pfreq__ (becuase of correlation with "clt" at -0.67 -> "clt" has higher correlation with target)
- __title_mr__ (becuase of correlation with "title_mrs" at -0.93)
- __unsized_dummy__ (becuase of corrlation with "woe_item_size" at 0.69 -> "woe_item_size" has higher correlation with target)
- __color_keeper__ (becuase of corrlation with "woe_item_color" at 0.53 -> "woe_item_color" has higher correlation with target)
- __ref. price variables__ (mean_ref_price, 7d_rolling_mean_ref_price, 7d_rolling_median_ref_price, max_ref_price) 


```python
features_to_drop = [
    'most_exp_item'
    , 'rel_pfreq'
    , 'title_mr'
    , 'unsized_dummy'
    , 'color_keeper'
    , 'mean_ref_price'
    , '7d_rolling_mean_ref_price'
    , '7d_rolling_median_ref_price'
    , 'max_ref_price'
]
known.drop(labels=features_to_drop, axis=1, inplace=True)
unknown.drop(labels=features_to_drop, axis=1, inplace=True)
```

__Next__:
1. plot feature importances (using RF-classifier in sklearn)
2. do recursive feature elimnation (using sklearn with RF-classifier)

```python
from sklearn.ensemble import RandomForestClassifier

arg_dict = {
    'n_estimators': 100
    , 'max_depth': 4
    , 'random_state': 2022
    , 'max_samples': 0.7
}
clf = RandomForestClassifier(max_depth=4, random_state=2022)
clf.fit(known.loc[:,~known.columns.isin(['order_item_id', 'return'])], known['return'])
```

```python
fi_df = pd.DataFrame({
    'features': clf.feature_names_in_
    , 'importances': clf.feature_importances_
}).sort_values(by='importances',ascending=False).reset_index(drop=True)

fig,ax = plt.subplots(1,1,figsize=(12,8))

sns.barplot(data=fi_df,y='features',x='importances',ax=ax)
plt.show()
```

## 3. Recursive Feature Elimination

```python
from sklearn.feature_selection import RFECV

arg_dict = {
    'n_estimators': 100
    , 'max_depth': 4
    , 'random_state': 2022
    , 'max_samples': 0.7
}
clf = RandomForestClassifier(**arg_dict)

selector = RFECV(clf, step=1, cv=5)
selector = selector.fit(known.loc[:,~known.columns.isin(['order_item_id', 'return'])], known['return'])
```

```python
pd.DataFrame({
    'features':[col for col in known.columns if col not in ['order_item_id', 'return']]
    , 'rank': selector.ranking_ 
}).sort_values(by='rank').reset_index(drop=True)
```
